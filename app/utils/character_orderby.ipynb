```python
html = pydocx.PyDocX.to_html('data/sample.docx')
soup = BeautifulSoup(html)
```

```python
p_list = soup.select('p')  #标签选择器
print(len(p_list))
flag = 0
character_list = []
img_list = []
for p in p_list:
    if flag%2 == 0:
        character_list.append(p.get_text())
    else:
        img_list.append(str(p))
    flag += 1
```

```python
def cmp_fisrt_4(x,y):
    try:
        x_i = int(x[0][:4])
        y_i = int(y[0][:4])
        if x_i>y_i:
            return True
        else:
            return False
    except:
        return True
```

```python
cp = []
for i in range(0, len(img_list)):
    cp.append((character_list[i], img_list[i]))
```

```python
a = sorted(cp,key=lambda x:x[0])
print(len(a))
```

```python
f_out = open("data/out.html","w",encoding="utf8")
for item in a:
    print("<p>",end="",file=f_out)
    print(item[0],end="",file=f_out)
    print("</p>",file=f_out)
    print(item[1],file=f_out)
f_out.close()
```
