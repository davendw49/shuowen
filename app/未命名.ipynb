```python
import base64
from docx import Document
```

```python
value = '<img height="33px" src="" width="38px"/>'
img = base64.b64decode(value)
file = open('test.jpg','wb') 
file.write(img)  
file.close()
```

```python
document = Document()
```

```python
document.styles['Normal'].font.name = u'黑体'
document.add_heading("doc", 0)
document.add_picture("test.jpg")
document.save("test.docx")
```
